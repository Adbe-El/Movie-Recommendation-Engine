```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
movie_ratings = pd.merge(movies, ratings)
```

```python
movie_ratings.head()
```

```python
#Select Useful columns
movie_ratings.drop(['genres', 'timestamp'],axis=1, inplace=True)
```

```python
movie_ratings.head()
```

```python
user_rating = movie_ratings.pivot_table(index='userId', columns='title', values='rating')
user_rating.head()
```

```python
# Select only cases where NAN rating equals 10 and below
user_rating = user_rating.dropna(thresh=10).fillna(0)
user_rating.head()
```

```python
#Let's Build our Similarity Matrix

item_similarity = user_rating.corr(method='pearson')
item_similarity.head()
```

```python
#Let's Make A Recommendations

def get_similar_movies(movie_name, user_rating):
    similar_score = item_similarity[movie_name]* (user_rating - 2.5)
    similar_score = similar_score.sort_values(ascending=False)
    
    return similar_score
```

```python
#For the case when a user inputs more than one movie name and rating
movies_ratings = [("'Hellboy': The Seeds of Creation (2004)"", 2), ('Salem\'s Lot (2004)', 3), ('(500) Days of Summer (2009)',4)]

movie_recommendation = pd.DataFrame()

for movie, rating in movies_ratings:
    movie_recommendation = movie_recommendation.append(get_similar_movies(movie, rating), ignore_index=True)
    
movie_recommendation.sum().sort_values(ascending=False)
```
